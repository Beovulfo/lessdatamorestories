```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('../data/cladribine.csv', sep=';')
```

```python
df = df.sort_values(by='Persistence (median HR)')
```

```python
df['Treatment'] = df['Treatment'].str.strip()
```

```python
import matplotlib.pyplot as plt

# Set custom colors
colors = ['#800080', '#FFFF00', '#00FF00']

# Set custom font
font = {'family': 'Arial', 'weight': 'normal', 'size': 14}

# Set the figure and axes
fig, ax = plt.subplots()

# Example data
treatments = df['Treatment'].values
y_pos = np.arange(len(treatments))
performance = df['Persistence (median HR)'].values
error = df['Persistence (half IQR)'].values

# Create the bar plot with custom colors
ax.barh(y_pos, performance, xerr=error, align='center', color=colors)

# Set y-axis labels
ax.set_yticks(y_pos)
ax.set_yticklabels(treatments, fontdict=font)

# Invert y-axis
ax.invert_yaxis()

# Set x-axis label
ax.set_xlabel('Treatment Persistence (HR)', fontdict=font)

# Set plot title
#ax.set_title('Treatment Persistence', fontdict=font)

# Show the plot
plt.show()
```

```python
df = pd.read_csv('../data/paciente.csv', sep=';')
df["ao"] = df['ao'].astype(int)
df["Coste anual"] = df['Coste anual'].apply(lambda x: x/1000)
df["Coste paciente"] = df['Coste paciente'].apply(lambda x: x/1000)
colors =['#51087E', '#808080', '#90EE90']
# Set the figure and axes
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(10, 8), sharex="col")

# Plot the first subplot (Coste anual)
ax1.plot(df['ao'], df['Coste anual'], color=colors[1], marker='o', linestyle='-', linewidth=2)

# Get the current x-axis tick labels
#xticks = ["{:d}".format(v) for v in xticks]

#ax1.set_xlabel('Ao ', fontdict=font, fontsize=fontsize)
ax1.set_ylabel('Coste total(x1000)', fontdict=font, fontsize=fontsize)

ax1.set_xticklabels(ax1.get_xticklabels(), 
                    font=basier_square_mono,
                    rotation=0,
                    size=fontsize)

last_point = len(df) - 1
ax1.annotate(f'{df["Coste anual"].values[last_point]:.0f}k', 
              xy=(df['ao'].values[last_point]-0.2, df['Coste anual'].values[last_point]-100),
              #xytext=(df['ao'].values[last_point], df['Coste anual'].values[last_point] + -100),
              #arrowprops=dict(facecolor='black', arrowstyle='->'),
              color=colors[1],
              font=basier_square_mono,
              fontsize=fontsize)

#Inceament
#
last_change = (df['Coste anual'].values[-1] - df['Coste anual'].values[-2])/df['Coste anual'].values[-2] * 100
ax1.annotate(f'{last_change:.0f}%', 
              xy=(df['ao'].values[last_point]-0.5, (df['Coste anual'].values[last_point]
                  +df['Coste anual'].values[last_point-1])/2.0),
              #xytext=(df['ao'].values[last_point], df['Coste anual'].values[last_point] + -100),
              #arrowprops=dict(facecolor='black', arrowstyle='->'),
              color=colors[0],
              font=basier_square_mono,
              fontsize=fontsize)
#

last_point = len(df) - 2
ax1.annotate(f'{df["Coste anual"].values[last_point]:.0f}k', 
              xy=(df['ao'].values[last_point]-0.2, df['Coste anual'].values[last_point]-100),
              #xytext=(df['ao'].values[last_point], df['Coste anual'].values[last_point] + -100),
              #arrowprops=dict(facecolor='black', arrowstyle='->'),
              color=colors[1],
              font=basier_square_mono,
              fontsize=fontsize)


# Plot the first subplot (Coste anual)
ax2.plot(df['ao'], df['Coste paciente'], color=colors[1], marker='o', linestyle='-', linewidth=2)

# Get the current x-axis tick labels
#xticks = ["{:d}".format(v) for v in xticks]

ax2.set_xlabel('Ao ', fontdict=font, fontsize=fontsize)
ax2.set_ylabel('Coste paciente(x1000)', fontdict=font, fontsize=fontsize)
#ax2.set_xticks()



last_point = len(df) - 1
ax2.annotate(f'{df["Coste paciente"].values[last_point]:.0f}k', 
              xy=(df['ao'].values[last_point]-0.1, df['Coste paciente'].values[last_point]+1),
              #xytext=(df['ao'].values[last_point], df['Coste anual'].values[last_point] + -100),
              #arrowprops=dict(facecolor='black', arrowstyle='->'),
              color=colors[1],
              font=basier_square_mono,
              fontsize=fontsize)

last_change = (df['Coste paciente'].values[-1] - df['Coste paciente'].values[-2])/df['Coste paciente'].values[-2] * 100
ax2.annotate(f'{last_change:.0f}%', 
              xy=(df['ao'].values[last_point]-0.5, (df['Coste paciente'].values[last_point]
                  +df['Coste paciente'].values[last_point-1])/2.0),
              #xytext=(df['ao'].values[last_point], df['Coste anual'].values[last_point] + -100),
              #arrowprops=dict(facecolor='black', arrowstyle='->'),
              color=colors[0],
              font=basier_square_mono,
              fontsize=fontsize)
#

last_point = len(df) - 2
ax2.annotate(f'{df["Coste paciente"].values[last_point]:.0f}k', 
              xy=(df['ao'].values[last_point], df['Coste paciente'].values[last_point]+0.5),
              #xytext=(df['ao'].values[last_point], df['Coste anual'].values[last_point] + -100),
              #arrowprops=dict(facecolor='black', arrowstyle='->'),
              color=colors[1],
              font=basier_square_mono,
              fontsize=fontsize)

ax1.set_ylim([0,800])
ax2.set_ylim([0,20])
ax1.set_yticklabels(ax1.get_yticklabels(),
                   font=basier_square_mono,
                   rotation=0,
                   size=fontsize)
ax2.set_yticks([0,5,10,15,20])
ax2.set_yticklabels(ax2.get_yticklabels(),
                   font=basier_square_mono,
                   rotation=0,
                   size=fontsize)


ax2.set_xticks(df['ao'].values)

ax2.set_xticklabels(df['ao'].values, 
                    font=basier_square_mono,
                    rotation=0,
                    size=fontsize)
# Adjust the layout
plt.tight_layout()

# Show the plot
plt.show()
```
