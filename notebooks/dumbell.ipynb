```python
import plotly.graph_objects as go
from plotly import data

def plot_dumbell_country(df_plot, values, colors, country):
    """
    Plot dumbell for given country
    """

    df_country = df_plot[df_plot["country"]==country]
    categories = (
        df_country["category_name"].unique()
    )

    data = {"line_x": [], "line_y": [], values[0]: [], values[1]:[], values[2]: [], "colors": [], 
        "categories":[],
        "countries": []}

    for category in categories:
        starting_point = df_country.loc[
            (df_country.category_name == category) & (df_country.step==values[0])]["mean"].values[0]*100
    
        middle_point = df_country.loc[
            (df_country.category_name == category) & (df_country.step==values[1])]["mean"].values[0]*100
    
        end_point = df_country.loc[
            (df_country.category_name == category) &(df_country.step==values[2]) ]["mean"].values[0]*100

        data[values[0]].extend([starting_point])
        data[values[1]].extend([middle_point])
        data[values[2]].extend([end_point])

        data["line_x"].extend(
            [
                starting_point,
                end_point,
                None,
            ]
        )
        data["line_y"].extend([category, category, None]),

    fig = go.Figure(
        data=[
            go.Scatter(
                x=data["line_x"],
                y=data["line_y"],
                mode="lines",
                showlegend=False,
                marker=dict(
                    color="grey"
                )
            ),
            go.Scatter(
                x=data[values[0]],
                y=categories,
                mode="markers",
                name=values[0],
                marker=dict(
                    color=colors[0],
                    size=10
                )
            ),
            go.Scatter(
                x=data[values[1]],
                y=categories,
                mode="markers",
                name=values[1],
                marker=dict(
                    color=colors[1],
                    size=10
                )   
            ),

            go.Scatter(
                x=data[values[2]],
                y=categories,
                mode="markers",
                name=values[2],
                marker=dict(
                    color=colors[2],
                    size=10
                )   
            ),
        ]
    )

    fig.update_layout(
        title=f"Evolution for {country}",
        height=1000,
        legend_itemclick=False
    )

    fig.show()
```
